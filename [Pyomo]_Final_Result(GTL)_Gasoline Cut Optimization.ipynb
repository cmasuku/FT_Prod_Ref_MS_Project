```python
# Given kinetic constants by Pellegrini et al.
mu5 = 1/3               
mu5star = 27/96
mu6 = 29/30
#--------------------------#
# Kliso [MPa -1] 
Kliso59 = 5.46e0 *1e6
Kliso1014 = 5.48e1 *1e6
Kliso1522 = 9.46e2 *1e6
Kliso22 = 9.53e7 *1e6
#--------------------------#
# Kln [Mpa-1 ]
Kln14 = 5.46e-2 * 1e6
Kln59 = 5.47e0 * 1e6
Kln1014 = 5.48e1 * 1e6
Kln1522 = 9.05e2 * 1e6
Kln22 = 7.19e7 * 1e6
#--------------------------#
# Keq [Dimensionless]
Keq59 = 7.19e2
Keq1014 = 8.36e2
Keq1522 = 3.76e3
Keq22 = 5.73e3
#--------------------------#
# k0iso [kmol/(sec*kg_cat)]
k0iso59 = 1.14e23 / 3600
k0iso1014 = 2.60e24 / 3600
k0iso1522 = 9.47e27 / 3600
k0iso22 = 2.80e29 / 3600
#--------------------------#
# k0cr [kmol/(sec*kg_cat)]
k0cr59 = 4.48e22 / 3600             
k0cr1014 = 9.50e24 / 3600
k0cr1522 = 7.05e26 / 3600                   
k0cr22 = 7.90e27 / 3600
#--------------------------#
# Eiso [kJ/kmol]
Eiso59 = 1.94e5
Eiso1014 = 1.94e5
Eiso1522 = 1.94e5
Eiso22 = 2.06e5
#--------------------------#
# Ecr [kJ/kmol]
Ecr59 = 1.30e5
Ecr1014 = 1.66e5
Ecr1522 = 1.82e5
Ecr22 = 1.87e5
#--------------------------#
T0 =T= 375+273.15  # Temperature at inlet [K]
P0 = 4.75 #6.96  # Pressure inlet [Mpa] 
#--------------------------#
# Physical Condition
length = 15 
diameter = 2.5   # Reactor diameter[m]
Area = np.pi * (diameter/2)**2. # Reactor cross section area [m^2]
R = 8.314  # kJ/(kmol*K)
#--------------------------#
# Total flows at inlet
Vin = 0.0409061543436171 # 0.041 #0.00292       # Actual volume flow [m3/s] 
#--------------------------#
mu_gas = 2.5140e-05 * 1e-6   # Dynamic viscosity [Mpa*s]
#--------------------------#
# Catalyst 
rhocat = 950  # kg/m3 reactor  
epsilon = 0.8 # Void fraction
dp = 0.005 # diameter of Catalyst particle   
rhogas = 8.6843 # gas density [kg/m3]
vs = Vin / Area # Superficial gas velocity [m/s]
G = vs*rhogas   # kg/(m^3*s)
Rep = G*dp/mu_gas # partical Raynolds number
f = 150 + 1.75 * Rep/(1-epsilon) # friction factor
#--------------------------#
```

# Weight Fraction of Hydrocarbons and H2/Wax Ratio 

```python
Wax = 6.475 # kg/sec 

H2_Wax = 0.105  

# weight fraction of wax 
WFC14 = 1e-5
WFC59 = 1e-5
WFC1014 = 1e-5
WFC1522 = 1e-5
WFC2356 = 1

# MW [Kg/Kmol]
MWH2 = 2.016 # Kg/Kmol 
MWC14 = 37.0825 # Kg/Kmol
MWC59 = 84 # Kg/Kmol 
MWC1014 = 150 # Kg/Kmol 
MWC1522 = 210 # Kg/Kmol 
MWC2356 = 350 # Kg/Kmol 

WC14 = Wax * WFC14 # kg/sec
WC59 = Wax * WFC59 # kg/sec
WC1014 = Wax * WFC1014 # kg/sec
WC1522 = Wax * WFC1522 # kg/sec
WC2356 = Wax * WFC2356 # kg/sec
WH2 = Wax* H2_Wax # kg/sec

MC14 = WC14 / MWC14 # kmol/sec
MC59 = WC59 / MWC59 # kmol/sec
MC1014 = WC1014 / MWC1014 # kmol/sec
MC1522 = WC1522 / MWC1522 # kmol/sec
MC2356 = WC2356 / MWC2356 # kmol/sec
MH2 = WH2 / MWH2 # kmol/sec

print(f'Mass flow rate of H2 is {WH2} kg/sec')
print(f'Mass flow rate of C9- is {WC14} kg/sec')
print(f'Mass flow rate of C9- is {WC59} kg/sec')
print(f'Mass flow rate of C1014 is {WC1014} kg/sec')
print(f'Mass flow rate of C1522 is {WC1522} kg/sec')
print(f'Mass flow rate of C2370 is {WC2356} kg/sec')
print()
print(f'Molar flow rate of H2 is {MH2:1.5f} kmol/sec.')
print(f'Molar flow rate of C9- is {MC14:1.5f} kmol/sec.')
print(f'Molar flow rate of C9- is {MC59:1.5f} kmol/sec.')
print(f'Molar flow rate of C1014 is {MC1014:1.5f} kmol/sec.')
print(f'Molar flow rate of C1522 is {MC1522:1.5f} kmol/sec.')
print(f'Molar flow rate of C2370 is {MC2356:1.5f} kmol/sec.')
```

```python
def k1(T): # isomerization of nC2370
    return k0iso22 * np.exp(-Eiso22/R/T)
def k2(T): # isomerization of nC1522
    return k0iso1522 * np.exp(-Eiso1522/R/T)
def k3(T): # isomerization of nC1014
    return k0iso1014 * np.exp(-Eiso1014/R/T)
def k4(T): # isomerization of nC59
    return k0iso59 * np.exp(-Eiso59/R/T)
#-----------------------------------------------#
def k5(T): # hydrocracking of iso C2370
    return k0cr22 * np.exp(-Ecr22/R/T)
def k6(T): # hydrocracking of iso C1522
    return k0cr1522 * np.exp(-Ecr1522/R/T)
def k7(T): # hydrocracking of iso C1014
    return k0cr1014 * np.exp(-Ecr1014/R/T)
def k8(T): # hydrocracking of iso C59
    return k0cr59 * np.exp(-Ecr59/R/T)
#-----------------------------------------------#
```

# ODE Function 

```python
def dFdZ(F,Z): # ode function 
    
    Fh2,Fco,Fco2,Fh20,FC14,FC59,FC1014,FC1522,FC2370,FiC59,FiC1014,FiC1522,FiC2370,y = F
    P = y*P0
    Ft = Fh2+Fco+Fco2+Fh20+FC14+FC59+FC1014+FC1522+FC2370+FiC59+FiC1014+FiC1522+FiC2370
    
    Ph2 = Fh2 / Ft * P   # patial pressure of h2
    Pco = Fco / Ft * P    # patial pressure of co
    Pco2 = Fco2 / Ft * P  # patial pressure of co2
    Ph20 = Fh20 / Ft * P  # patial pressure of h20
    PC14 = FC14 / Ft * P  # patial pressure of C14
    PC59 = FC59 / Ft * P  # patial pressure of C59
    PC1014 = FC1014 / Ft * P # patial pressure of C1014
    PC1522 = FC1522 / Ft * P # patial pressure of C1522
    PC2370 = FC2370 / Ft * P # patial pressure of C2370
    PiC59 = FiC59 / Ft * P    # patial pressure of iC59
    PiC1014 = FiC1014 / Ft * P  # patial pressure of iC1014
    PiC1522 = FiC1522 / Ft * P  # patial pressure of iC1522
    PiC2370 = FiC2370 / Ft * P  # patial pressure of iC2370


    Adsorb = Ph2*(1+Kln22*PC2370+Kliso22*PiC2370+Kln1522*PC1522+Kliso1522*PiC1522
                  +Kln1014*PC1014+Kliso1014*PiC1014+Kln59*PC59+Kliso59*PiC59+Kln14*PC14)
    
    if Ph2 > 0: # reaction rates [kmol/(kgcat*sec)]
        #Isomerization    
        r1 = k1(T)*(PC2370-(PiC2370/Keq22))/Adsorb # kmol/(kgcat*sec)
        r2 = k2(T)*(PC1522-(PiC1522/Keq1522))/Adsorb # kmol/(kgcat*sec)
        r3 = k3(T)*(PC1014-(PiC1014/Keq1014))/Adsorb # kmol/(kgcat*sec)
        r4 = k4(T)*(PC59-(PiC59/Keq59))/Adsorb # kmol/(kgcat*sec)
        #Hydrocracking
        r5 = k5(T)*PiC2370/Adsorb # kmol/(kgcat*sec)
        r6 = k6(T)*(PiC1522)/Adsorb # kmol/(kgcat*sec)
        r7 = k7(T)*PiC1014/Adsorb # kmol/(kgcat*sec)
        r8 = k8(T)*PiC59/Adsorb # kmol/(kgcat*sec)
   
    else:  # converge when all hydrogen is consumed
        r1 = 0                                      
        r2 = 0
        r3 = 0
        r4 = 0
        r5 = 0
        r6 = 0
        r7 = 0
        r8 = 0
                   

    # Rate expressions for each species 
    rn2370 = -r1 # kmol/(kgcat*sec)
    riso2370 = r1 + (-1 + 2*(1-mu5-mu5star))*r5 # kmol/(kgcat*sec)   
    rn1522 = -r2 # kmol/(kgcat*sec)
    riso1522 = r2 + 2*mu5*r5 -0.4*r6 # kmol/(kgcat*sec)        
    rn1014 = -r3 # kmol/(kgcat*sec)        
    riso1014 = r3 + 2*mu5star*r5 + 2*(1-mu6)*r6 - r7 # kmol/(kgcat*sec)   
    rn59 = -r4 # kmol/(kgcat*sec)
    riso59 = r4 + 2*mu6*r6 + 2*r7 - r8 # kmol/(kgcat*sec)
    rn14 = 2*r8 # kmol/(kgcat*sec)
    
    
    
    
    # (kmol/hr)/m --> dF/dZ: flow rate change over reactor length   
    dFdZn1 = rhocat * (1-epsilon) * rn14 * Area # nC14 [(kmol/sec)/m]
    dFdZn2 = rhocat * (1-epsilon) * rn59 * Area # nC59 [(kmol/sec)/m]
    dFdZn3 = rhocat * (1-epsilon) * rn1014 * Area  # nC1014 [(kmol/sec)/m]
    dFdZn4 = rhocat * (1-epsilon) * rn1522 * Area  # nC1522 [(kmol/sec)/m]
    dFdZn5 = rhocat * (1-epsilon) * rn2370 * Area  # nC2370 [(kmol/sec)/m]
    dFdZi1 = rhocat * (1-epsilon) * riso59 * Area   # iC59 [(kmol/sec)/m]
    dFdZi2 = rhocat * (1-epsilon) * riso1014 * Area  # iC1014 [(kmol/sec)/m]
    dFdZi3 = rhocat * (1-epsilon) * riso1522 * Area  # iC1522 [(kmol/sec)/m]
    dFdZi4 = rhocat * (1-epsilon) * riso2370 * Area  # iC2370 [(kmol/sec)/m]
    #dFdZH2 = rhocat * (1-epsilon) * rH2 * Area # H2 [(kmol/sec)/m]
    dFdZH2 = -(dFdZn1+dFdZn2+dFdZn3+dFdZn4+dFdZn5+dFdZi1+dFdZi2+dFdZi3+dFdZi4)
    dFdZCO = 0 # CO [(kmol/sec)/m]
    dFdZCO2 = 0 # CO2 [(kmol/sec)/m]
    dFdZH2O = 0 # H20 [(kmol/sec)/m]
                                                                              
    # 1/m  dy/dZ: nondemensionalized pressure change over reactor length
    # Pressure equation (Ergun's equation)            
    dy = -1/(P0*1e6)*f*G*mu_gas/(rhogas*dp**2)*(1-epsilon)**2/epsilon**3
    return [dFdZH2,dFdZCO, dFdZCO2, dFdZH2O, dFdZn1, dFdZn2, dFdZn3, dFdZn4, dFdZn5, dFdZi1, dFdZi2, dFdZi3, dFdZi4, dy]
```

# Solve 

```python
from scipy.integrate import odeint
```

```python

python_pts = 20 
lspan = np.linspace(0,length,python_pts)
#-----------------------------------------------#
# initial conditions for odeint 
b0 = [MH2,  0, 0, 0,  MC14,   MC59 ,MC1014, MC1522, MC2356,  0,  0,  0,  0, 1]

#-----------------------------------------------#
sol = odeint(dFdZ,b0,lspan)
#-----------------------------------------------#

# collect results of differential variables
Fh2 = sol[:, 0]
Fco = sol[:, 1]
Fco2 = sol[:, 2]
Fh20 = sol[:, 3]
FC14 = sol[:,4]
FC59 = sol[:,5]
FC1014 = sol[:,6]
FC1522 = sol[:, 7]
FC2370 = sol[:, 8]
FiC59 = sol[:,9]
FiC1014 = sol[:,10]
FiC1522 = sol[:,11]
FiC2370 = sol[:,12]
y = sol[:,13]

P = y*P0
Ftotal = FC14+FC59+FC1014+FC1522+FC2370+FiC59+FiC1014+FiC1522+FiC2370+Fh2+Fco+Fco2+Fh20 # total flow rate results

Ph2 = Fh2 / Ftotal * P   # patial pressure of h2
Pco = Fco / Ftotal * P    # patial pressure of co
Pco2 = Fco2 / Ftotal * P  # patial pressure of co2
Ph20 = Fh20 / Ftotal * P  # patial pressure of h20
PC14 = FC14 / Ftotal * P  # patial pressure of C14
PC59 = FC59 / Ftotal * P  # patial pressure of C59
PC1014 = FC1014 / Ftotal * P # patial pressure of C1014
PC1522 = FC1522 / Ftotal * P # patial pressure of C1522
PC2370 = FC2370 / Ftotal * P # patial pressure of C2370
PiC59 = FiC59 / Ftotal * P    # patial pressure of iC59
PiC1014 = FiC1014 / Ftotal * P  # patial pressure of iC1014
PiC1522 = FiC1522 / Ftotal * P  # patial pressure of iC1522
PiC2370 = FiC2370 / Ftotal * P  # patial pressure of iC2370

Adsorb = Ph2*(1+Kln22*PC2370+Kliso22*PiC2370+Kln1522*PC1522+Kliso1522*PiC1522
              +Kln1014*PC1014+Kliso1014*PiC1014+Kln59*PC59+Kliso59*PiC59+Kln14*PC14)

r1 = k0iso22 * np.exp(-Eiso22/R/T)*(PC2370-(PiC2370/Keq22))/Adsorb
r2 = k0iso1522 * np.exp(-Eiso1522/R/T)*(PC1522-(PiC1522/Keq1522))/Adsorb
r3 = k0iso1014 * np.exp(-Eiso1014/R/T)*(PC1014-(PiC1014/Keq1014))/Adsorb
r4 = k0iso59 * np.exp(-Eiso59/R/T)*(PC59-(PiC59/Keq59))/Adsorb        
r5 = k0cr22 * np.exp(-Ecr22/R/T)*PiC2370/Adsorb
r6 = k0cr1522 * np.exp(-Ecr1522/R/T)*(PiC1522)/Adsorb
r7 = k0cr1014 * np.exp(-Ecr1014/R/T)*PiC1014/Adsorb
r8 = k0cr59 * np.exp(-Ecr59/R/T)*PiC59/Adsorb


# Rate expressions for each species 
rn2370 = -r1 # kmol/(kgcat*hr)
riso2370 = r1 + (-1 + 2*(1-mu5-mu5star))*r5 # kmol/(kgcat*hr)   
rn1522 = -r2 # kmol/(kgcat*hr)
riso1522 = r2 + 2*mu5*r5 -0.4* r6 # kmol/(kgcat*hr)        
rn1014 = -r3 # kmol/(kgcat*hr)        
riso1014 = r3 + 2*mu5star*r5 + 2*(1-mu6)*r6 - r7 # kmol/(kgcat*hr)   
rn59 = -r4 # kmol/(kgcat*hr)
riso59 = r4 + 2*mu6*r6 + 2*r7 - r8 # kmol/(kgcat*hr)
rn14 = 2*r8 # kmol/(kgcat*hr)
    
dFdZn1 = rhocat * (1-epsilon) * rn14 * Area             
dFdZn2 = rhocat * (1-epsilon) * rn59 * Area              
dFdZn3 = rhocat * (1-epsilon) * rn1014 * Area           
dFdZn4 = rhocat * (1-epsilon) * rn1522 * Area              
dFdZn5 = rhocat * (1-epsilon) * rn2370 * Area              
dFdZi1 = rhocat * (1-epsilon) * riso59 * Area              
dFdZi2 = rhocat * (1-epsilon) * riso1014 * Area            
dFdZi3 = rhocat * (1-epsilon) * riso1522 * Area            
dFdZi4 = rhocat * (1-epsilon) * riso2370 * Area  

dFdZH2 = -(dFdZn1+dFdZn2+dFdZn3+dFdZn4+dFdZn5+dFdZi1+dFdZi2+dFdZi3+dFdZi4)

                                                     
                                                           
#  results of mole fraction of each components
Ftotal_hydrocarbons = FC14+FC59+FC1014+FC1522+FC2370+FiC59+FiC1014+FiC1522+FiC2370 # total flow rate results

X_inC14 = FC14/Ftotal_hydrocarbons
X_inC59 = FC59/Ftotal_hydrocarbons
X_inC1014 = FC1014/Ftotal_hydrocarbons
X_inC1522 = FC1522/Ftotal_hydrocarbons
X_inC2370 = FC2370/Ftotal_hydrocarbons
X_iniC59 = FiC59/Ftotal_hydrocarbons
X_iniC1014 = FiC1014/Ftotal_hydrocarbons
X_iniC1522 = FiC1522/Ftotal_hydrocarbons
X_iniC2370 = FiC2370/Ftotal_hydrocarbons
#-----------------------------------------------#
dspan = lspan / length * 100
#-----------------------------------------------#
```

# Define Variables 

```python
model = m = ConcreteModel()

# model reactor parameters
model.L = Param(initialize = 15) # total reactor length = 50m   Param : fixed variable 
    
model.z = ContinuousSet(bounds = (0,model.L)) # z direction length  
model.Species = Set(initialize = ['H2','CO','CO2','H20','C14','C59','C1014','C1522','C2370','iC59',
                                  'iC1014','iC1522','iC2370' ])

model.Reactions = RangeSet(8) # number of Reaction : 8 Reactions 
model.P0 = Var(initialize = 4.75, bounds = (3.5,5.00)) # Pressure [Mpa] 

model.T =  Var(initialize = 375 + 273.15) # Tempature [K] 

# Physics 
model.diameter = Param(initialize = 2.5) # Reactor diameter[m]
model.Area = Param(initialize = np.pi * (m.diameter/2)**2.) # Reactor cross section area [m^2]
model.R = Param(initialize = 8.314) # kJ/(kmol*K)

model.Vf_in = Var(initialize = 0.041, bounds =(Vin_lb, Vin_up)) # Actual volume flow [m3/s] 
model.mu_gas = Param(initialize = 2.5140e-05 * 1e-6 ) # Dynamic viscosity [Mpa*s]
model.rho_gas = Param(initialize = 8.6843)  # gas density [kg/m3]
model.H2_Wax_ratio = Var(initialize = 0.105, bounds = (1e-2,1.5))

model.Rep = Var(initialize = Rep) # Reynold number of Catalyst particle  
model.f = Var(initialize = f)  # friction constant

# Catalyst 
model.rhocat = Param(initialize = 950) # kg/m3 reactor  
model.epsilon = Param(initialize = 0.8)  # Void fraction
model.dp = Param(initialize = 0.005) # diameter of catalyst particle

# define Variables 
model.F = Var(model.z, model.Species,  initialize = 0.0185) # molar flow rate [kg/sec]
model.Ft = Var(model.z, initialize = 0.5717) # Total molar flow rate [kg/sec]
model.Pt = Var(model.z,initialize = model.P0) #, bounds=(4.75,4.75)) # Pressure [Mpa]...P[0] from python 
model.P = Var(model.z, model.Species, initialize = model.P0) # Pressure [Mpa]
model.y = Var(model.z, initialize = 1, bounds = (0.1,1))#within = NonNegativeReals)
model.Adsorb = Var(model.z, initialize = Adsorb[0]) 
model.rate = Var(model.z, model.Reactions, initialize = r1[0])
model.k = Var(model.z, model.Reactions, initialize = 1) 

# define derivative variables 
model.dF = DerivativeVar(model.F, wrt = model.z, initialize = 1e-3) 
model.dy = DerivativeVar(model.y, wrt = model.z, initialize = 1e-3) 
```

# H2/Wax Ratio

```python
model.Wax = Param(initialize =  6.475) # kg/sec of wax  



# MW [Kg/Kmol]
model.MWH2 = Param(initialize = 2.016)  # Kg/Kmol 

model.MoleH2 = Var(initialize = 0.33724)
model.MoleC14 = Param(initialize = 1e-5)
model.MoleC59 = Param(initialize  = 1e-5)
model.MoleC1014 = Param(initialize  = 1e-5)
model.MoleC1522 = Param(initialize  = 1e-5)
model.MoleC2356 = Param(initialize  = 0.0185)
```

# Kinetic / Energy Parameter (Param)

```python
### Given kinetic constants by Pellegrini et al.
#-----------------------------------------#
model.mu5 = Param(initialize = 1/3)               
model.mu5star = Param(initialize =27/96)
model.mu6 = Param(initialize = 29/30)
#-----------------------------------------#
# Kliso [Mpa-1]
model.Kliso59 = Param(initialize = 5.46e0 * 1e6)
model.Kliso1014 = Param(initialize = 5.48e1 * 1e6)
model.Kliso1522 = Param(initialize = 9.46e2 * 1e6)
model.Kliso22 = Param(initialize = 9.53e7 * 1e6)
#-----------------------------------------#
# Kln [Mpa]
model.Kln14 = Param(initialize = 5.46e-2 * 1e6)
model.Kln59 = Param(initialize = 5.47e0 * 1e6)
model.Kln1014 = Param(initialize = 5.48e1 * 1e6)
model.Kln1522 = Param(initialize = 9.05e2 * 1e6)
model.Kln22 = Param(initialize = 7.19e7 * 1e6)
#-----------------------------------------#
# Keq [Dimensionless]
model.Keq59 = Param(initialize = 7.19e2)
model.Keq1014 = Param(initialize = 8.36e2)
model.Keq1522 = Param(initialize = 3.76e3)
model.Keq22 = Param(initialize = 5.73e3)
#-----------------------------------------#
# k0iso [kmol/(sec*kg_cat)]
model.k0iso59 = Param(initialize = 1.14e23/3600)
model.k0iso1014 = Param(initialize = 2.60e24/3600)
model.k0iso1522 = Param(initialize = 9.47e27/3600)
model.k0iso22 = Param(initialize = 2.80e29/3600)
#-----------------------------------------#
# k0cr [kmol/(sec*kg_cat)]
model.k0cr59 = Param(initialize = 4.48e22/3600)             
model.k0cr1014 = Param(initialize = 9.50e24/3600)
model.k0cr1522 = Param(initialize = 7.05e26/3600)                   
model.k0cr22 = Param(initialize = 7.90e27/3600)
#-----------------------------------------#
# Eiso [kJ/kmol]
model.Eiso59 = Param(initialize = 1.94e5)
model.Eiso1014 = Param(initialize = 1.94e5)
model.Eiso1522 = Param(initialize = 1.94e5)
model.Eiso22 = Param(initialize = 2.06e5)
#-----------------------------------------#
# Ecr [kJ/kmol]
model.Ecr59 = Param(initialize = 1.30e5)
model.Ecr1014 = Param(initialize = 1.66e5)
model.Ecr1522 = Param(initialize = 1.82e5)
model.Ecr22 = Param(initialize = 1.87e5)
```

# Constraints

```python
def Con_Total_flowrate_rule(m,z):
    return m.Ft[z] == sum(m.F[z,i] for i in model.Species)
model.Con_Total_flowrate = Constraint(m.z, rule= Con_Total_flowrate_rule)

def Con_Rep_rule(m):
    return m.Rep == m.Vf_in / m.Area * m.rho_gas * m.dp / m.mu_gas
model.Con_Rep = Constraint(rule = Con_Rep_rule)

def Con_f_rule(m):
    return m.f == 150 + 1.75 * m.Rep/(1-m.epsilon)
model.Con_f = Constraint(rule = Con_f_rule)

def Con_dy_rule(m,z): # this can be in the params 
    return m.dy[z] == -1 / (m.P0*1e6)*m.f*m.Vf_in / m.Area*m.rho_gas* m.mu_gas/(m.rho_gas*m.dp**2) * (1-m.epsilon)**2/m.epsilon**3
model.Con_dy = Constraint(model.z, rule = Con_dy_rule)

def Con_Pressure_drop_rule(m,z):
    return m.Pt[z] == m.y[z]* m.P0
model.Con_pressure_drop = Constraint(model.z, rule = Con_Pressure_drop_rule)

def Con_Partial_pressure_rule(m,z,i): # partial pressure of each components 
    return m.P[z,i] * m.Ft[z] == m.F[z,i]* m.Pt[z]
model.Con_Partial_pressure = Constraint(m.z, model.Species, rule= Con_Partial_pressure_rule)

def Con_Adsorb_rule(m,z):    
    return m.Adsorb[z] == m.P[z,'H2']*(1 + m.Kln22*m.P[z,'C2370']+ m.Kliso22 * m.P[z,'iC2370'] + m.Kln1522 * m.P[z,'C1522'] 
                                       + m.Kliso1522 * m.P[z,'iC1522'] + m.Kln1014 * m.P[z,'C1014'] + m.Kliso1014* m.P[z,'iC1014']
                                       + m.Kln59 * m.P[z,'C59'] + m.Kliso59 * m.P[z,'iC59'] + m.Kln14 * m.P[z,'C14']) 
model.Con_Adsorb = Constraint(m.z, rule = Con_Adsorb_rule)

def Con_Molarflowrate_H2_rule(m):
    return m.MoleH2 == m.Wax * m.H2_Wax_ratio / m.MWH2
model.Con_Molarflowrate_H2 = Constraint(rule= Con_Molarflowrate_H2_rule)
```

```python
def Con_k1_rule(m,z):
    return m.k[z,1] == m.k0iso22 * exp(-m.Eiso22/m.R/m.T)
model.Con_k1 = Constraint(m.z, rule = Con_k1_rule)

def Con_k2_rule(m,z):
    return m.k[z,2] == m.k0iso1522 * exp(-m.Eiso1522/m.R/m.T)
model.Con_k2 = Constraint(m.z, rule = Con_k2_rule)

def Con_k3_rule(m,z):
    return m.k[z,3] == m.k0iso1014 * exp(-m.Eiso1014/m.R/m.T)
model.Con_k3 = Constraint(m.z, rule = Con_k3_rule)

def Con_k4_rule(m,z):
    return m.k[z,4] == m.k0iso59 * exp(-m.Eiso59/m.R/m.T)
model.Con_k4 = Constraint(m.z, rule = Con_k4_rule)

def Con_k5_rule(m,z):
    return m.k[z,5] == m.k0cr22 * exp(-m.Ecr22/m.R/m.T)
model.Con_k5 = Constraint(m.z, rule = Con_k5_rule)

def Con_k6_rule(m,z):
    return m.k[z,6] == m.k0cr1522 * exp(-m.Ecr1522/m.R/m.T)
model.Con_k6 = Constraint(m.z, rule = Con_k6_rule)

def Con_k7_rule(m,z):
    return m.k[z,7] == m.k0cr1014 * exp(-m.Ecr1014/m.R/m.T)
model.Con_k7 = Constraint(m.z, rule = Con_k7_rule)

def Con_k8_rule(m,z):
    return m.k[z,8] == m.k0cr59 * exp(-m.Ecr59/m.R/m.T)
model.Con_k8 = Constraint(m.z, rule = Con_k8_rule)
```

## Rate equtions

```python
def Con_rate1_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,1]== 0
    else:
        return m.rate[z,1] * m.Adsorb[z] == m.k[z,1]*(m.P[z,'C2370']-m.P[z,'iC2370']/m.Keq22)         
model.Con_rate1 = Constraint(m.z, rule = Con_rate1_rule)

def Con_rate2_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,2]== 0
    else:
        return m.rate[z,2] * m.Adsorb[z] ==  m.k[z,2]*(m.P[z,'C1522']-m.P[z,'iC1522']/m.Keq1522) 
model.Con_rate2 = Constraint(m.z, rule = Con_rate2_rule)

def Con_rate3_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,3]== 0
    else:
        return m.rate[z,3] * m.Adsorb[z] == m.k[z,3]*(m.P[z,'C1014']-m.P[z,'iC1014']/m.Keq1014) 
model.Con_rate3 = Constraint(m.z, rule = Con_rate3_rule)

def Con_rate4_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,4]== 0
    else:
        return m.rate[z,4] * m.Adsorb[z] == m.k[z,4]*(m.P[z,'C59']-m.P[z,'iC59']/m.Keq59)
model.Con_rate4 = Constraint(m.z, rule = Con_rate4_rule)

def Con_rate5_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,5]== 0
    else:
        return m.rate[z,5] * m.Adsorb[z] == m.k[z,5]*(m.P[z,'iC2370'])
model.Con_rate5 = Constraint(m.z, rule = Con_rate5_rule)

def Con_rate6_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,6]== 0
    else:
        return m.rate[z,6] * m.Adsorb[z]  == m.k[z,6]*(m.P[z,'iC1522']) 
model.Con_rate6 = Constraint(m.z, rule = Con_rate6_rule)

def Con_rate7_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,7]== 0
    else:
        return m.rate[z,7] * m.Adsorb[z]== m.k[z,7]*(m.P[z,'iC1014']) 
model.Con_rate7 = Constraint(m.z, rule = Con_rate7_rule)

def Con_rate8_rule(m,z):
    if m.P[z,'H2'] == 0:
        return m.rate[z,8]== 0
    else:
        return m.rate[z,8] * m.Adsorb[z] == m.k[z,8]*(m.P[z,'iC59']) 
model.Con_rate8 = Constraint(m.z, rule = Con_rate8_rule)
```

## ODE Constraints

```python
# Reaction Systems 
def Con_FC14_rule(m,z):
    return m.dF[z,'C14'] == m.rhocat * (1-m.epsilon) * m.Area * 2 * m.rate[z,8]
m.Con_FC14 = Constraint(m.z, rule = Con_FC14_rule)

def Con_FC59_rule(m,z):
    return m.dF[z,'C59'] == m.rhocat * (1-m.epsilon) * m.Area * -1 * m.rate[z,4]
m.Con_FC59 = Constraint(m.z, rule = Con_FC59_rule)

def Con_FC1014_rule(m,z):
    return m.dF[z,'C1014'] == m.rhocat * (1-m.epsilon) * m.Area * -1 * m.rate[z,3]
m.Con_FC1014 = Constraint(m.z, rule = Con_FC1014_rule)

def Con_FC1522_rule(m,z):
    return m.dF[z,'C1522'] == m.rhocat * (1-m.epsilon) * m.Area * -1 * m.rate[z,2]
m.Con_FC1522 = Constraint(m.z, rule = Con_FC1522_rule)

def Con_FC2370_rule(m,z):
    return m.dF[z,'C2370'] == m.rhocat * (1-m.epsilon) * m.Area * -1 * m.rate[z,1]
m.Con_FC2370 = Constraint(m.z, rule = Con_FC2370_rule)

def Con_FiC59_rule(m,z):  
    return m.dF[z,'iC59'] == m.rhocat * (1-m.epsilon) * m.Area * (m.rate[z,4] + 2 * m.mu6 * m.rate[z,6] + 2 * m.rate[z,7] - m.rate[z,8])
m.Con_FiC59 = Constraint(m.z, rule = Con_FiC59_rule)

def Con_FiC1014_rule(m,z):      
    return m.dF[z,'iC1014'] == m.rhocat * (1-m.epsilon) * m.Area * (m.rate[z,3] + 2 * m.mu5star * m.rate[z,5] 
                                                              + 2 * (1-m.mu6) * m.rate[z,6] - m.rate[z,7])
m.Con_FiC1014 = Constraint(m.z, rule = Con_FiC1014_rule)

def Con_FiC1522_rule(m,z):   
    return m.dF[z,'iC1522'] == m.rhocat * (1-m.epsilon) * m.Area * (m.rate[z,2] + 2 * m.mu5 * m.rate[z,5] - 0.4* m.rate[z,6])
m.Con_FiC1522 = Constraint(m.z, rule = Con_FiC1522_rule)

def Con_FiC2370_rule(m,z):
    return m.dF[z,'iC2370'] == m.rhocat * (1-m.epsilon) * m.Area * (m.rate[z,1] + (-1 + 2*(1-m.mu5-m.mu5star)) * m.rate[z,5])
m.Con_FiC2370 = Constraint(m.z, rule = Con_FiC2370_rule)

def Con_FH2_rule(m,z):  
    return m.dF[z,'H2'] == -(m.dF[z,'C14']+ m.dF[z,'C59']+ m.dF[z,'C1014']+ m.dF[z,'C1522']+ m.dF[z,'C2370']+ m.dF[z,'iC59']+ m.dF[z,'iC1014']+ m.dF[z,'iC1522']+ m.dF[z,'iC2370'])
m.Con_FH2 = Constraint(m.z, rule = Con_FH2_rule)

def Con_FCO_rule(m,z):  
    return m.dF[z,'CO'] == 0
m.Con_FCO = Constraint(m.z, rule = Con_FCO_rule)

def Con_FCO2_rule(m,z):  
    return m.dF[z,'CO2'] == 0
m.Con_FCO2 = Constraint(m.z, rule = Con_FCO2_rule)

def Con_FH20_rule(m,z):  
    return m.dF[z,'H20'] == 0
m.Con_FH20 = Constraint(m.z, rule = Con_FH20_rule)
```
